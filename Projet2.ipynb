```python
import pandas as pd 
# Charger le fichier Excel contenant les données "sales"
df = pd.read_excel(r"C:\Users\Cellphone\Downloads\donnees_maladies_sales.xlsx")
# Affichage des premières lignes pour voir à quoi ça ressemble
print("Données brutes : ")
df
```

```python
# ...............1. Supprimer les doublons ................................
# Supprimer les dublons(si un enregistrement est répété)
df = df.drop_duplicates()
df
```

```python
# ..............2. Traiter les valeurs maquantes .........................
# Affichages des valeurs maquantes par colonne 
print("\nValeurs manquantes par colonne : ")
print(df.isnull().sum())
```

```python
# Option 1 : Remplir les valeurs manquantes avec une valeur par défaut(par exemple, 0 pour les cas signalés,"Inconnu" pour noms
df = df.drop_duplicates()
df['Cas signalés']= df['Cas signalés'].fillna(0)
df['Nom'] = df['Nom'].fillna('Inconnu')
df['Sexe'] = df['Sexe'].fillna('Inconnu')
df
```

```python
# Option 2: Supprimer les lignes où les valeurs sont manquantes dans des colonnes essentielles comme 'Date' ou 'Maladie'
df = df.dropna(subset=['Date','Maladie'])
#.........................3. Correction des formats .......................
# Corriger les dates mal fiormatées 
df['Date']=pd.to_datetime(df['Date'],errors='coerce')#Cela va convertir les dates invalides en NaT(Not a Time)
df=df.dropna(subset=['Date']) # Cela supprime les lignes avec les dates invalides
# .......................4. Correction des majuscules/miniscules ...................
#Normaliser les noms de colonnes pour avoir la première lettre en majuscule(par exemple"M" ou "F" dans la colonne sexe)
df['Sexe'] = df['Sexe'].str.upper() # Met tout en majuscule pour avoir un format plus cohérent

# .......................5. Correction des valeurs abérantes.....................
#Filtrer les âges incohérentes(par exemple un âge de -5 ou 999)
df=df[(df['Âge']>=0) & (df['Âge']<=100)] # On garde uniquement les ages >=0 et <=100

# ......................6. Vérification du nettoyage..............................
df
```

```python
# Vérification des valeurs manquantes après nettoyage

print("\nValeurs manquantes après nettoyage : ")
print(df.isnull().sum())
```

```python
# Sauvegarder le fichier nettoyé dans un nouveau fichier Excel 
df.to_excel("donnees_malades_nettoyees.xlsx",index=False)
```
